```python
import numpy as np
import random as rnd
import itertools
```

### 1. Написать функцию формирования порождающей и проверочной матриц кода Хэмминга $$ ({2}^{r}-1, {2}^{r}-r-1, 3) $$ на основе параметра *r*, а также таблицы синдромов для всех однократных ошибок

```python
def get_G_H_matrix(n, k, d):
    x_matr = init_x_matrix(n - k, k, d)
    eye_matr_G = np.eye(x_matr.shape[0], dtype=int)
    eye_matr_H = np.eye(x_matr.shape[1], dtype=int)
    return np.concatenate((eye_matr_G, x_matr), axis=1), np.concatenate((x_matr, eye_matr_H), axis=0)
```

```python
def init_x_matrix(n, k, d):
    """
    n - число столбцов (длина строки)
    k - число строк
    d - минимальное число единиц в строке -(минус) 1
    """
    x_matr = []
    while (len(x_matr) != k):
        row = [0 for t in range(n)]
        for _ in range(n):
            flag = rnd.randint(0, 1)
            if (flag):
                idx = rnd.randint(0, n - 1)
                row[idx] = 1
            if (is_correct_row(x_matr, row, d) and not row in x_matr):
                x_matr.append(row)
                break
    return np.array(x_matr)
```

```python
def is_correct_row(x_matr, row, d):
    X = x_matr.copy()
    X.append(row)
    flag = True
    for i in range(1, d):
        comb = list(itertools.permutations(X, i))
        for c in comb:
            sum_ = np.sum(list(c), axis=0) % 2
            if sum_.tolist().count(1) < (d - i):
                flag = False
                break
        if not flag:
            break
    return flag
```

```python
def get_syndrome(h_matr):
    rows = h_matr.shape[0]
    eye_matr = np.eye(rows, dtype=int)
    syndroms_list = np.dot(eye_matr, h_matr) % 2
    syndrome_dict = dict()
    for i in range(len(syndroms_list)):
        syndrome_dict[tuple(syndroms_list[i])] = eye_matr[i, :]
    return syndrome_dict
```

### 2. Провести исследование кода Хэмминга для одно-, двух- и трёхкратных ошибок для  *r* = 2, 3, 4

```python
def get_err_word(g_matr, h_matr, syndrome, t):
    row = g_matr.shape[0]
    idx = rnd.randint(0, row - 1)
    u = np.array(g_matr[idx][:row])
    v = np.dot(u, g_matr) % 2
    print(f"Исходное сообщение: {u}")
    print(f"Отправленное сообщение: {v}", )
    v_err = v.copy()
    for i in range(t):
        v_err[i] += 1
        v_err[i] %= 2
    print(f"Принятое сообщение с ошибкой: {v_err}")
    print(f"Декодированное сообщение без исправления: {v_err[:row]}")
    syndrom = np.dot(v_err, h_matr) % 2
    print(f"Синдром: {syndrom}")
    try:
        error = syndrome[tuple(syndrom)]
        print(f"Ошибка: {error}")
        fixed_message = (v_err + error) % 2
        print(f"Исправленное сообщение: {fixed_message}")
        print(f"Декодированное сообщение: {fixed_message[:row]}")
    except:
        print("Для такого синдрома не найдено ошибки")
```

```python
r_list = [2, 3, 4]
t_list = [1, 2, 3]
```

```python
for r in r_list:
    for t in t_list:
        print(f"\033[1mИсследование кода Хэмминга при r = {r} и t = {t}\033[0m")
        g_matr, h_matr = get_G_H_matrix(2 ** r - 1, 2 ** r - r - 1, 1)
        syndrome = get_syndrome(h_matr)
        get_err_word(g_matr, h_matr, syndrome, t)
        if (r != r_list[-1] or t != t_list[-1]):
            print("\n")
```

### 3. Написать функцию формирования порождающей и проверочной матриц расширенного кода Хэмминга $$ ({2}^{r}-1, {2}^{r}-r-1, 3) $$ на основе параметра *r*, а также таблицы синдромов для всех однократных ошибок

```python
def get_ext_G_matr(x_matr):
    x_rows = x_matr.shape[0]

    eye_matr_G = np.eye(x_rows, dtype=int)
    g_matr = np.concatenate((eye_matr_G, x_matr), axis=1)
    b = np.array([g_matr[i].sum() % 2 for i in range(x_rows)]).reshape((x_rows, 1))

    return np.concatenate((g_matr, b), axis=1)
```

```python
def get_ext_H_matr(x_matr):
    eye_matr_H = np.eye(x_matr.shape[1], dtype=int)
    h_matr = np.concatenate((x_matr, eye_matr_H), axis=0)
    j = np.ones((h_matr.shape[0], 1) , dtype = int)
    h_ext_by_j_matr = np.concatenate((h_matr, j), axis=1)
    last_row = np.zeros((1,h_ext_by_j_matr.shape[1]), dtype = int)
    last_row[0][-1] = 1

    return np.concatenate((h_ext_by_j_matr, last_row), axis=0)
```

```python
def get_ext_G_H_matrix(n, k, d):
    x_matr = init_x_matrix(n - k, k, d)

    return get_ext_G_matr(x_matr), get_ext_H_matr(x_matr)
```

### 4. Провести исследование расширенного кода Хэмминга для одно-, двух-, трёх- и четырёхкратных ошибок для *r* = 2, 3, 4

```python
for r in r_list:
    for t in t_list:
        print(f"\033[1mИсследование расширенного кода Хэмминга при r = {r} и t = {t}\033[0m")
        g_ext_matr, h_ext_matr = get_ext_G_H_matrix(2 ** r - 1, 2 ** r - r - 1, 1)
        syndrome = get_syndrome(h_ext_matr)
        get_err_word(g_ext_matr, h_ext_matr, syndrome, t)
        if (r != r_list[-1] or t != t_list[-1]):
            print("\n")
```
